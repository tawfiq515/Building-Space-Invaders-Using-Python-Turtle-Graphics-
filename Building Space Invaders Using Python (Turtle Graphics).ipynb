```python
import turtle
import random

screen = turtle.Screen()
screen.title("Space Invaders")
screen.bgcolor("black")
screen.setup(width=800, height=600)

border = turtle.Turtle()
border.color("white")
border.penup()
border.goto(-300, -250)
border.pendown()
border.forward(600)
border.hideturtle()

player = turtle.Turtle()
player.shape("triangle")
player.color("white")
player.penup()
player.goto(0, -230)
player.setheading(90)

def move_left():
    x = player.xcor()
    if x > -280:
        player.setx(x - 20)

def move_right():
    x = player.xcor()
    if x < 280:
        player.setx(x + 20)

screen.listen()
screen.onkey(move_left, "Left")
screen.onkey(move_right, "Right")

aliens = []
for i in range(5):
    alien = turtle.Turtle()
    alien.shape("circle")
    alien.color("red")
    alien.penup()
    alien.goto(random.randint(-250, 250), 200)
    aliens.append(alien)

bullet = turtle.Turtle()
bullet.shape("square")
bullet.color("yellow")
bullet.shapesize(0.2, 1)
bullet.penup()
bullet.hideturtle()
bullet_state = "ready"

def fire_bullet():
    global bullet_state
    if bullet_state == "ready":
        bullet_state = "fired"
        bullet.goto(player.xcor(), player.ycor())
        bullet.showturtle()

def move_bullet():
    global bullet_state
    if bullet_state == "fired":
        bullet.sety(bullet.ycor() + 15)
        if bullet.ycor() > 250:
            bullet.hideturtle()
            bullet_state = "ready"

alien_speed = 0.2
def move_aliens():
    for alien in aliens:
        alien.sety(alien.ycor() - alien_speed)
        if alien.ycor() < -230:
            screen.bye()  

def check_collision():
    global bullet_state
    for alien in aliens:
        if bullet.distance(alien) < 20:
            alien.goto(random.randint(-250, 250), 200)
            bullet.hideturtle()
            bullet_state = "ready"

screen.onkey(fire_bullet, "space")


while True:
    screen.update()
    move_bullet()
    move_aliens()
    check_collision()
```
